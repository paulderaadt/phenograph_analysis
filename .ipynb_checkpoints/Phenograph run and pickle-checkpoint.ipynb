```python
import phenograph
import pandas as pd
import numpy as np
import community
import networkx as nx
from sklearn.neighbors import kneighbors_graph
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
#import scanpy
from matplotlib.colors import Normalize
from matplotlib.patches import Patch
```

```python
data = pd.read_csv("../../LDA/Data/AML_benchmark.csv", header=0)
```

```python
print(data.shape)
```

Prepare sample by removing nongated cells and nonmarker attributes

```python
data = data.loc[data['cell_type'] != 'NotDebrisSinglets']

data = data.drop(axis='columns', labels=["Time", "Cell_length", "DNA1", "DNA2",
                                       "Viability", "file_number", "event_number", "subject"])
datalabels = pd.DataFrame(data['cell_type'])
labels = data['cell_type']
data = data.drop(axis='columns', labels="cell_type")
data.head()
```

Arcsin transform

```python
data = np.arcsinh((data-1)/5)
```

```python
data.shape
```

Call phenograph

```python

communities, graph, Q = phenograph.cluster(data.values)
```

```python
unique, counts = np.unique(communities, return_counts=True)
dict(zip(unique, counts))
```

Create kNN graph from sample data

```python
nxgraph = nx.Graph(graph)
```

```python
phenograph_communities = pd.DataFrame(counts, unique)
phenograph_communities.columns = ['count']
phenograph_communities['community'] = phenograph_communities.index
```

Call Louvain community method without phenograph

```python
louvain_communities = community.best_partition(nxgraph)
```

```python
louvainpd = pd.DataFrame(pd.Series(louvain_communities))
louvainpd.columns = ['community']
louvainpd['node'] = louvainpd.index
louvainpd = pd.DataFrame(louvainpd.groupby("community").count()).sort_values(by=['node'], ascending=False)
```

```python
louvainpd['community'] = louvainpd.index

```

```python
#tsne from file
raw_embedded = pd.read_csv("./TSNE.csv", header=-1)
raw_embedded.columns = ['x', 'y']
x = raw_embedded['x'].values
y = raw_embedded['y'].values
```

```python
plt.rcParams['figure.figsize'] = [15, 8]
```

```python
labelset = list(set(labels))
colors = []
norm = Normalize(vmin=0, vmax=13)
cmap = plt.cm.get_cmap('Spectral')
for i in range(14):
    colors.append(cmap(norm(i)))
```

```python
fig, ax = plt.subplots(1, 1)
ax.scatter(x,y, s=2, c=[colors[labelset.index(i)] for i in labels])
plt.xlabel("TSNE 1")
plt.ylabel("TSNE 2")
handles = []
for i in range(14):
    handles.append(Patch(facecolor=colors[i]))
plt.legend(handles, labelset, loc='upper right', prop={'size': 7})
plt.title("TSNE of gated labels")
plt.show()
```

```python
data = pd.read_csv("../../LDA/Data/AML_benchmark.csv", header=0)
data = data.loc[data['cell_type'] != 'NotDebrisSinglets']

data = data.drop(axis='columns', labels=["Time", "Cell_length", "DNA1", "DNA2",
                                       "Viability", "file_number", "event_number", "subject"])

data.head()
```

```python
from sklearn.model_selection import ShuffleSplit
```

```python
splitter = ShuffleSplit(n_splits=1, test_size=0.5, random_state=0)

for train, test in splitter.split(data):
    traindata = data.iloc[train]
    testdata = data.iloc[test]
testdata = testdata.drop(columns='cell_type')
```

```python
c, p = phenograph.classify(test=testdata.values, train=traindata.values)
```
